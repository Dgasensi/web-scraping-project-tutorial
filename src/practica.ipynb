## Scraping Tesla Quarterly revenue

```python
#url del sitio a scrapear
url = " https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

#headers definidos para evitar errores en el request
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36"}

#efectuamos el request del codigo html
response = requests.get(url, headers=headers)
#definimos un condicional para manejar errores de importación
if response.status_code != 200:
    print('error al conectar')
#convertimos la respuesta a texto
response_html = response.text
```

```python
#creamos el objeto soup 
soup = bs.BeautifulSoup(response_html, 'html.parser')
#buscamos todos los elementos de la clase 'table'
tablas = soup.findAll(class_ = 'table')
#selecciono el indice 1 de la lista 'tablas' que se corresponde con lo que busco
tablas_Tesla_Quarterly = tablas[1]
#extraigo todas las filas de la lista
filas = tablas_Tesla_Quarterly.tbody.findAll('tr')

#defino dos variables para almacenar los datos de las filas
valores = []
fechas = []
#Iteramos sobre cada una de las filas para extraer los valores de las columnas 'fecha' y 'valor'
for tr in filas:
    columna = tr.findAll('td')
    fecha = columna[0].text
    valor = columna[1].text
    fechas.append(fecha)
    valores.append(valor)

#creamos una variable para almacenar los valores numericos limpios    
valores_limpios = []
#iteramos sobre la lista valores para eliminar espacios, simbolos y convertir a entero
for valor in valores:
    valor_limpio = valor.replace('$', '')
    valor_limpio = valor_limpio.replace(',', '').strip()
    #comprobamos que el valor existe y lo añadimos a la lista
    if valor_limpio:
        valores_limpios.append(int(valor_limpio))
    else:
        valores_limpios.append(None)
        
#creamos el dataframe
df = pd.DataFrame({'fecha': fechas, 'valor': valores_limpios})
df
```

```python
#url del sitio a scrapear
url = 'https://es.wikipedia.org/wiki/Anexo:Elementos_qu%C3%ADmicos'
#realizamos la peticion al sitio y convertimos a texto
response = requests.get(url)
response_html = response.text
#defino condicional para manejo de excepcion
if response.status_code != 200:
    print('error al conectar')

# Creamos objeto soup
soup = bs.BeautifulSoup(response_html, 'lxml')
# Buscamos la tabla que nos interesa
table = soup.find('table', {'class': 'wikitable center'})

```

```python
#creamos las listas donde se almacenaran cada uno de los datos de cada columna
numeros_atomicos = []
simbolos = []
elementos = []
pesos_atomicos = []
densidades = []
puntos_fusion = []
puntos_ebullicion = []
electronegatividades = []
abundancias = []
links = []

# Iteramos sobre cada fila desde el indice [2] para evitar el encabezado y una fila vacia//probramos el modulo tqdm(barra de progreso)
for row in tqdm(table.findAll('tr')[2:]):

    # Almacenamos en la variable tds todos los valores de cada fila
    tds = row.findAll('td')

    # Seleccionamos el texto de cada valor que queremos de la lista tds (los indices han sido escogidos previamente) 
    numero_atomico = tds[0].text
    simbolo = tds[1].text
    elemento = tds[2].text
    peso_atomico = tds[5].text
    densidad = tds[6].text
    fusion = tds[7].text
    ebullicion = tds[8].text
    electronegatividad = tds[10].text
    abundancia = tds[11].text
    etimologia = tds[14].text
    # Extraemos tambien el enlace al elemento 
    link = tds[2].find('a')['href'] if tds[2].find('a') else None

    # Añadimos cada valor a su respectiva lista 
    numeros_atomicos.append(numero_atomico)
    simbolos.append(simbolo)
    elementos.append(elemento)
    pesos_atomicos.append(peso_atomico)
    densidades.append(densidad)
    puntos_fusion.append(fusion)
    puntos_ebullicion.append(ebullicion)
    electronegatividades.append(electronegatividad)
    abundancias.append(abundancia)
    links.append(link)

# Ahora tenemos qeu modificar los links para incluir el dominio general ya que en la tabla solo teniamos la ruta relativa
links_completos = []
dominio = 'https://es.wikipedia.org'
# Iteramos sobre la lista conlos links
for link in links:
    # Añadimos a cada link el dominio general
    link_completo = dominio + link
    # Añadimos los links ya completados a la nueva lista [links_completos]
    links_completos.append(link_completo)
```

```python
''' UN INTENTO DE LIMPIAR LOS DATOS NUMERICOS QUE NO HA FUNCIONADO PORQUE EL PATRON NO ES CORRECTO Y ELIMINABA MUCHOS TERMINOS'''
import re
# detras de exclamacion/con o sin espacio/busca digito + decimal opcional (?:)antes de corchetes
patron = r'!\s*(\d+(?:\.\d+)?)[\[\]]'
patron2 = r'(\d+(?:\.\d+)?)'

pesoA_limpio = [re.search(patron, texto).group(1)  if re.search(patron, texto) else None for texto in pesos_atomicos]
fusion_limpio = [re.search(patron2, texto).group(1) if re.search(patron2, texto) else None for texto in puntos_fusion ]
```
